Проведем обработку данных:

```python
# Drop the last column without any data
data = data.drop(data.columns[-1], axis=1)
# Drop the `id` column
data = data.drop('id', axis=1)
# Convert the diagnosis column to numeric format
data['diagnosis'] = data['diagnosis'].factorize()[0]
# Fill all NaN values with zero
data = data.fillna(value=0)
# Store the diagnosis column in a target object and then drop it
target = data['diagnosis']
data = data.drop('diagnosis', axis=1)

data
```

Выполним нормализацию полученных данных:

```python
from sklearn import preprocessing

data = pd.DataFrame(
    preprocessing.MinMaxScaler().fit_transform(data),
    columns=data.columns,
)
data
```

## Задание 2

Выполнить визуализацию многомерных данных, используя t-SNE.
Рассмотреть результаты визуализации для разных значений перплексии.

Определим списки значений параметров, которые пригодятся в будущем:

```python
# Random state for all algorithms
random_state = 42
# Define required parameters
n_neighbours = [5, 25, 50, 100,]
min_dists = [0.1, 0.6,]
```

```python
import matplotlib.pyplot as plt

%matplotlib notebook
```

Напишем код для получения результатов с помощью алгоритма t-SNE:

```python
from sklearn.manifold import TSNE

def calc_tsne(n_components: int, perplexity: int) -> pd.DataFrame:
    tsne = TSNE(n_components=n_components,
                perplexity=perplexity,
                random_state=random_state)
    return pd.DataFrame(tsne.fit_transform(data))
```

Напишем код для отображения результатов с помощью алгоритма t-SNE:

```python
def draw_tsne(perplexity: int, is3d: bool = False):
    dims = 3 if is3d else 2
    result = calc_tsne(dims, perplexity)
    projection = '3d' if is3d else None
    args = [result.iloc[:, i] for i in range(dims)]

    plt.figure()
    ax: plt.Axes = plt.axes(projection=projection)
    ax.scatter(*args, c=target, cmap='coolwarm', alpha=0.5)
    ax.set_title(f't-SNE algorithm results with perplexity of {perplexity}')
    ax.set_axisbelow(True)
    ax.grid(axis='both')
```

Отобразим двухмерные графики с учетом различных методов решения задачи классификации, а также значений перплексии:

```python
[draw_tsne(perplexity) for perplexity in n_neighbours]
None
```

Отобразим трехмерные графики с учетом различных методов решения задачи классификации, а также значений перплексии:

```python
[draw_tsne(perplexity, is3d=True) for perplexity in n_neighbours]
None
```

## Задание 3

Выполнить визуализацию многомерных данных, используя UMAP с различными параметрами `n_neighbors` и `min_dist`.

Напишем код для получения результатов с помощью алгоритма UMAP:

```python
from umap import UMAP

def calc_umap(n_components: int, n_neighbors: int, min_dist: float) -> pd.DataFrame:
    umap = UMAP(n_neighbors=n_neighbors, n_components=n_components,
                min_dist=min_dist, random_state=random_state)
    return pd.DataFrame(umap.fit_transform(data))
```

Напишем код для отображения результатов с помощью алгоритма UMAP:

```python
# noinspection SpellCheckingInspection
def draw_umap(neighbor: int, is3d: bool = False):
    dims = 3 if is3d else 2
    projection = '3d' if is3d else None

    fig: plt.Figure = plt.figure()
    fig.suptitle(f'UMAP algorithm results with neighbors of {neighbor}')
    axs = fig.subplots(1, 2, subplot_kw=dict(projection=projection))
    for ax, min_dist in zip(axs.flat, min_dists):
        result = calc_umap(dims, neighbor, min_dist)
        args = [result.iloc[:, i] for i in range(dims)]

        ax.scatter(*args, c=target, cmap='coolwarm', alpha=0.5)
        ax.set_title(f'min_dist={min_dist}')
        ax.set_axisbelow(True)
        ax.grid(axis='both')
```

Отобразим двухмерные графики с учетом различных методов решения задачи классификации, а также значений числа соседей и минимального расстояния между точками:

```python
[draw_umap(neighbor) for neighbor in n_neighbours]
None
```

Отобразим трехмерные графики с учетом различных методов решения задачи классификации, а также значений числа соседей и минимального расстояния между точками:

```python
[draw_umap(neighbor, is3d=True) for neighbor in n_neighbours]
None
```

## Задание 4

Рассчитать время работы алгоритма UMAP с помощью библиотеки `time` и сравнить его с временем работы t-SNE.

Создадим таблицу, в которую будем записывать результаты вычислений времени исполнения алгоритмов:

```python
import time

times = pd.DataFrame({
    'Число соседей': [pd.NA for _ in range(len(n_neighbours) - 1)]
}).append({
    '2D t-SNE': pd.NA, '2D UMAP md=0.1': pd.NA, '2D UMAP md=0.6': pd.NA,
    '3D t-SNE': pd.NA, '3D UMAP md=0.1': pd.NA, '3D UMAP md=0.6': pd.NA,
}, ignore_index=True)

for i, neighbor in enumerate(n_neighbours):
    times.at[i, 'Число соседей'] = neighbor

times
```

Напишем код для подсчета времени выполнения алгоритма t-SNE:

```python
for neighbor in n_neighbours:
    for n_components in range(2, 4):
        t = time.process_time()
        calc_tsne(n_components, neighbor)
        ix = n_neighbours.index(neighbor)
        times.at[ix, f'{n_components}D t-SNE'] = time.process_time() - t
```

Напишем код для подсчета времени выполнения алгоритма UMAP:

```python
for neighbor in n_neighbours:
    for n_components in range(2, 4):
        for min_dist in min_dists:
            t = time.process_time()
            calc_umap(n_components, neighbor, min_dist)
            ix = n_neighbours.index(neighbor)
            times.at[ix, f'{n_components}D UMAP md={min_dist}'] = time.process_time() - t
```

Отобразим таблицу с временем выполнения алгоритмов:

```python
times
```
